## **LADME - losy leków w organizmie**

Procesy, którymi zajmuje się farmakokinetyka można podzielić na różne etapy, począwszy od przyjęcia leku, aż do jego wydalenia. System, który je opisuje nazywa się LADME i pochodzi od pierwszych liter angielskich nazw poniższych przemian:




*   L – liberation (uwalnienie) 
*   A – absorption (wchłanianie)
*   D – distribution (rozmieszczenie)
*   M – metabolism (metabolizm)
*   E – excretion lub elimination (wydalanie)





**Uwalnianie** - przechodzenie substancji czynnej zawartej w leku do środowiska, z którego może zostać wchłonięta do organizmu.

**Wchłanianie** - przejście substancji leczniczej do krążenia ogólnego, na tym etapie występują największe straty.

**Rozmieszczenie** - inaczej dystrybucja, rozprowadzenie leku w organizmie, dotarcie do miejsca działania.

**Metabolizm** - inaczej biotransformacja, biochemiczny proces przemian leków w organizmie, w wyniku którego powstają metabolity.

**Wydalanie** - usunięcie leków i ich metabolitów z organizmu [1].

## **Drogi podawania leków**

Pierwszym, a zarazem najważniejszym czynnikiem, który ma wpływ na przebieg opisanego powyżej procesu jest forma leku oraz miejsce jego podania. Podstawowe drogi podawania to:
* podanie donaczyniowe: 
  - dożylnie,
  - dotętniczo;
* podanie pozanaczyniowe: 
  - doustnie,
  - na powierzchnię ciała,
  - donosowo,

  - poprzez inhalację.




Każdy z tych sposobów wyróżnia się inna specyfiką i ma wpływ na szybkość poźniejszych procesów z systemu LADME, dlatego też bardzo ważne jest, aby zastosować odpowiednią metodę do konkretnego przypadku w celu osiągnięcia jak najlepszych efektów. 



## **Wchłanianie leków**

Na tym etapie dochodzi do największych strat, ponieważ lek musi się przedostać przez pewną "barierę" w postaci na przykład błony biologicznej. 

Czynniki wpływające na szybkość wchłaniania to między innymi:
* postać leku, 
* ilość leku,
* droga podania,
* motoryka przewodu pokarmowego,
* przyjmowane jednocześnie z lekiem pokarmy lub inne leki,
* środowisko, w którym zachodzi wchłanianie [2].

## **Model jednokompartmentowy**

Aby ułatwić opis procesów zachodzących po podaniu leku, organizm dzieli się na określone obszary, czyli kompartmenty (część ustroju, w której lek pod względem 
kinetycznym zachowuje się tak samo), które niekoniecznie odzwierciedlają jego faktyczną budowę. Do matematycznej interpretacji wyniku badań farmakokinetycznych będzie potrzebne także przyjęcie odpowiedniego modelu. Najprostszym z nich jest model jednokompartmentowy, który zakłada natychmiastowe i równomierne rozmieszczanie się leku w organizmie traktowanym jako jeden kompartment [3]. Dzięki stałej objętości kompartmentu całkowita masa leku jest proporcjonalna do jego stężenia (wydalanie jest procesem pierwszego rzędu). 

Parametry występujące w modelu:
*   $k_{e}$ - stała szybkości eliminacji (wydalania),
*   $k_{a}$ - stała szybkości wchłaniania,
*   $C$ - stężenie leku we krwi w czasie $t$,
*  $C_0$ - początkowe stężenie leku,
*   $C_{max}$ - maksymalne stęzenie leku we krwi,
*   $V_{d}$ - objętość dystrybucji,
* $t_{0,5}$ - biologiczny okres półtrwania leku,
* $t_{max}$ - czas uzyskania maksymalnego stężenia leku w ustroju.

## **Model jednokompartmentowy - podanie donaczyniowe**

Podając lek donaczyniowo, cała jego ilość w chwili początkowej znajduje się w kompartmencie.  W  tym  też  momencie  lek  osiąga  swoje maksymalne lokalne stężenie,  które z czasem zacznie maleć  ze  względu  na  eliminację  leku z organizmu. Korzystając z równań różniczkowych pierwszego rzędu możemy wyznaczyć funkcję masy leku w organizmie od czasu. Wygląda ona następująco [3]:

\begin{equation}
\frac{\operatorname{d}\!{X}}{\operatorname{d}\!t} = -{k_e}{X\\}
\label{eq:rownanie1}
 \tag{1}
\end{equation}

Gdzie:
* $X(t)$ - masa leku w miejscu podania w chwili $t$, $X \ge 0$,
* $k_{e}$ - stała szybkości eliminacji, $k_e > 0$,
* $X(0) = X_{0}$ - masa podanego leku, $X_{0} > 0$.

To proste równanie rozwiązujemy metodą rozdzielania zmiennych.

\begin{equation}
\int \frac{\operatorname{d}\!{X}}{X} = \int -k_e \operatorname{d}\!t
\end{equation}

\begin{equation}
\ln X = -k_et + C
\end{equation}

\begin{equation}
X = De^{-k_et}
\end{equation}

Stałą $D$ wyznaczamy z warunku początkowego $X(0)=X_0$

\begin{equation}
X_0=D
\end{equation}

A więc ostatecznie nasze równanie opisujące eliminację leku dla modelu jednokomparmentowego przy podaniu donaczyniowym ma postać:

\begin{gather}
X = X_0e^{-k_et}
\label{eq:rownanie2}
 \tag{2}
\end{gather}

Wynika z tego, że ilość leku maleje wykładniczo, co widać na poniższym wykresie:

```python
import matplotlib.pyplot as plt
import numpy as np

k_e = 0.6
X_0 = 50

t = np.linspace(0,10,2000)
X = X_0 * np.exp(-k_e * t)
t_05 = np.log(2)/k_e

plt.plot(t, X, 'm')
plt.title("Zmiana masy leku przy podaniu donaczyniowym")
plt.xlabel("Czas [h]")
plt.ylabel("Masa leku [mg]")
plt.xlim(0, 10)
plt.ylim(0,50)
plt.yticks([0,10,20,30,40,50], [0,10,20,30,40, 'C_max'])
plt.xticks([0,4,6,8,10, t_05], ["t_max",4,6,8,10, "t_0,5"])
plt.axvline(t_05, linestyle = '--', color = 'black', ymin = 0, ymax = 0.5, linewidth = 1)
plt.show()
```

Zaprezentowane równanie nie jest zbytnio praktyczne, ponieważ wyznaczenie ilości leku w ustroju jest niemożliwe. Dlatego też wyznacza się nie masę leku, lecz jego stężenie. Służy do tego pojęcie objętości dystrybucji $V_d$, czyli hipotetyczna objętość płynów ustrojowych, w której lek po równomiernym rozmieszczeniu miałby takie stężenie, jak to zmierzone we krwi. Związek masy leku, jego stężenia oraz objętości dystrybucji wyraża się wzorem:

\begin{gather}
X = {V_d} * {C}
\tag{3}
\label{eq:rownanie3}
\end{gather}

Gdzie:
* $V_d$ - objętość dystrybucji,
* $C$ - stężenie leku we krwi.

Po wstawieniu $\eqref{eq:rownanie3}$ do wzoru $\eqref{eq:rownanie2}$ uzyskujemy następującą postać:

\begin{gather}
V_d  C = V_d  C_0  e^{-k_et}
\end{gather}

Dzieląc obie strony przez $V_d$ otrzymujemy zależność wykładniczą, na podstawie któej możemy śledzić zmiany stężenia leku we krwi:

\begin{gather}
C = C_0e^{-k_et}
\end{gather}

oraz po zlogarytmowaniu:

\begin{gather}
ln C=ln C_0 -k_et
\end{gather}

Korzystając z powyższego wzoru jesteśmy w stanie podać biologiczny okres półtrwania, czyli czas wymagany do eliminacji połowy zaabsorbowanej dawki leku [4]. 
Wyznaczymy go poprzez wstawienie stężenia równego połowie steżenia początkowego do równania na $t$:

\begin{gather}
t = \frac{\operatorname{ln}\!\frac{C_0}{C}}{k_e}
\end{gather}



\begin{gather}
C = \frac{C_0}{2}
\end{gather}

Wtedy:

\begin{gather}
t_{0,5} = \frac{\operatorname{ln}\!\frac{2C_0}{C_0}}{k_e}
\end{gather}

Po uproszczeniu:

\begin{gather}
t_{0,5} = \frac{ln2}{k_e}
\end{gather}



Jego wartość silnie zależy od własności fizykochemicznych leku i wynosi od kilku minut do kilkudziesięciu dni. Przykładowe dane można zobaczyć w poniższej tabeli [3]:

|Lek|$t_{0,5}$[h]|
|--|--|
|Aspiryna|0,2 - 0,5|
|Ibuprofen|1 - 3|
|Penicylina G| 0,5 - 1|
|Naproksen|9 - 22|
|Witamina D3|800 - 1000|
|Witamina C| 0,2 - 0,5|



Stała szybkości eliminacji można wyznaczyć z nachylenia prostej dla funkcji liniowej $f(t)=ln C$.

## **Model jednokompartmentowy z wchłanianiem**

Inną formą podania leku jest podanie pozanaczyniowe (najczęściej doustne) [3]. W tym modelu przewód pokarmowy traktujemy jako wstępny kompartment, w którym następuje wchłanianie do krążenia ogólnego, opisane poniższym równaniem. Zazwyczaj proces wchłaniania jest szybszy od eliminacji, więc możemy zapisać, że $k_e \neq k_a$. Pomijamy zwrotne przenikanie leku do przewodu pokarmowego. 


\begin{equation}
\frac{\operatorname{d}\!{X_A}}{\operatorname{d}\!t} = -{k_a}{X_A}
\label{eq:rownanie4} 
\tag{4}
\end{equation}

Jego rozwiązanie to $X_A = X_0e^{-k_at}$.

Równanie opisujące zmianę masy leku w kompartmencie przy podaniu doustnym ma postać:

\begin{equation}
\frac{\operatorname{d}\!{X}}{\operatorname{d}\!t} = k_a X_A -k_eX
\end{equation}

Gdzie:
* $k_a$ - stała szybkości wchłaniania,
* $k_e$ - stała szybkości eliminacji,
* $X(0) = 0$ - warunek początkowy.

Podstawiając rozwiązanie równania $\eqref{eq:rownanie4}$ do obecnego otrzymujemy:

\begin{equation}
X'+k_eX=k_aX_0e^{-k_at}
\label{eq:rownanie5} 
\tag{5}
\end{equation}

Aby rozwiązać powyższe równanie, znajdziemy najpierw rozwiązanie równania jednorodnego, korzystając ponownie z metody rozdzielania zmiennych.

\begin{equation}
X'+k_eX=0
\end{equation}

\begin{equation}
X = Be^{-k_et}
\end{equation}

Więc rozwiązanie równania $\eqref{eq:rownanie5}$ ma postać $ X = Be^{-k_et}+\phi$. 

Następnie przewidujemy, że $\phi=Ae^{-k_et}$ i liczymy pochodną
$\phi' = A'e^{-k_et}-k_eAe^{-k_et}$.

Podstawiamy to do równania $\eqref{eq:rownanie5}$ i wyliczamy współczynnik A.

\begin{gather}
A'e^{-k_et}-k_eAe^{-k_et}+k_eAe^{-k_et}=k_aX_0e^{-k_at}
\end{gather}\
\begin{gather}
A'e^{-k_et}=k_aX_0e^{-k_at}
\end{gather}\
\begin{gather}
A' = k_aX_0e^{t(k_e-k_a)}
\end{gather}\
\begin{gather}
A = \frac{k_aX_0}{k_e-k_a}e^{t(k_e-k_a)} + C
\end{gather}

Czyli rozwiązanie ogólne ma postać:

\begin{gather}
X = Be^{-k_et} + e^{-k_et}\left(\frac{k_aX_0}{k_e-k_a}e^{t(k_e-k_a)} + C\right)
\end{gather} \
Zapisując inaczej: \
\begin{gather}
X = e^{-k_et}\left(\frac{k_aX_0}{k_e-k_a}e^{t(k_e-k_a)} + G\right)
\end{gather} \
Wyznaczając G z warunku początkowego: \
\begin{gather}
X(0) = 0 = \frac{k_aX_0}{k_e-k_a} + G 
\end{gather} \
\begin{gather}
G = - \frac{k_aX_0}{k_e-k_a}
\end{gather}



A więc ostatecznie równanie na zmianę masy leku w kompartmencie w zależności od czasu ma postać:

\begin{gather}
X(t) = \frac{k_aX_0}{k_e-k_a}\left(e^{-k_at}-e^{-k_et}\right)
\end{gather}

lub w formie stężeniowej:

\begin{gather}
C(t)= \frac{k_aX_0}{(k_e-k_a)V_d}\left(e^{-k_at}-e^{-k_et}\right)
\label{eq:rownanie6} 
\tag{6}
\end{gather}

Istotnym parametrem w tym modelu jest czas, po którym stężenie leku we krwi osiąga maksymalną wartość [6]. Można go wyznaczyć różniczkując równanie opisujące zmiany stężenia leku we krwi i znajdując czas, dla którego $\frac{dC}{dt}=0$.


\begin{gather}
\frac{dC}{dt} = \frac{k_aX_0}{(k_e-k_a)V_d}\left((-k_a)e^{-k_at}+k_e(e^{-k_et})\right)
\end{gather}\
\begin{gather}
(-k_a)e^{-k_at}+k_e(e^{-k_et}) =0
\end{gather}\
\begin{gather}
e^{(k_a-k_e)t}=\frac{k_a}{k_e}
\end{gather}\
\begin{gather}
(k_a-k_e)t = ln\left(\frac{k_a}{k_e}\right)
\end{gather}\
\begin{gather}
t_{max} = \frac{ln\left(\frac{k_a}{k_e}\right)}{k_a-k_e}
\end{gather}

Wartość $C_{max}$ można wyliczyć wstawiając wyliczone $t_{max}$ do równania $\eqref{eq:rownanie6}$.

\begin{gather}
C_{max} = \frac{k_aX_0}{(k_e-k_a)V_d}\left(e^{\frac{-k_a}{k_a-k_e}ln\left(\frac{k_a}{k_e}\right)}- e^{\frac{-k_e}{k_a-k_e}ln\left(\frac{k_a}{k_e}\right)}\right)
\end{gather}\
\begin{gather}
C_{max} = \frac{k_aX_0}{(k_e-k_a)V_d}\left(\left(\frac{k_a}{k_e}\right)^{\frac{-k_a}{k_a-k_e}}-\left(\frac{k_a}{k_e}\right)^{\frac{-k_e}{k_a-k_e}}\right)
\end{gather}

Zależności te są bardzo istotne, ponieważ podając określoną dawke leku, trzeba pamiętać, aby jego stężenie otrzymane w organizmie osiągnęło dawkę leczniczą, jednocześnie nie przekraczając dawki toksycznej (tzw. okno terapeutyczne). W kategoriach rejestracji nowych leków, gdzie również farmakokinetyka jest wykorzystywana, istotna jest różnica między stężeniem toksycznym, a skutecznym - lek jest bezpieczny, jeśli ta różnica jest dostatecznie duża. 

Na wykresie przedstawiamy zmianę stęzenia leku we krwi w zależności od czasu, wraz z wyróżnionym stężeniem maksymalnym i oknem terapeutycznym, przy podaniu dawki 100mg doustnie o szybkim uwalnianiu ($k_a = 1.4$, $k_e=0.2$, $V_d = 60$), czyli w przypadku, gdy $k_a > k_e$. 

```python
k_a = 1.4
k_e = 0.2
X_0 = 100
V_d = 60
toks = 1.5
min = 0.7

t = np.linspace(0, 20, 2000)
C = ((k_a*X_0)/((k_e - k_a)*V_d)) * (np.exp(-k_a*t) - np.exp(-k_e*t))

t_max = np.log(k_a/k_e) / (k_a - k_e)
C_max = ((k_a*X_0)/((k_e - k_a)*V_d)) * ((k_a/k_e)**(-k_a/(k_a-k_e))- (k_a/k_e)**(-k_e/(k_a-k_e)))

plt.plot(t, C, 'g')
plt.axhline(toks, linestyle = '--', color = 'black')
plt.axhline(min, linestyle = '--', color = 'black')
plt.axvline(t_max, linestyle = '--', color = 'black', ymin = 0, ymax = 0.6, linewidth = 1)
plt.axhline(y = C_max, linestyle = '--', color = 'black', xmin = 0, xmax = 0.083, linewidth = 1)
plt.xlim(0, 20)
plt.ylim(0, 2)
plt.yticks([toks, min, 1, 2, C_max], ['toks', 'min', 1, 2, 'C_max'])
plt.xticks([0,5,10,20,t_max], [0,5,10,20, 't_max'])
plt.title("Zmiana stężenia leku przy podaniu doustnym")
plt.xlabel("Czas [h]")
plt.ylabel("Stężenie [mg/l]")
plt.show()
```

Z wykresu możemy zobaczyć, że początkowo stężenie leku we krwi rośnie (lek się wchłania). W chwili $t_{max}$,  stężenie osiąga wartość maksymalną $C_{max}$, a następnie zaczyna maleć. Zakończenie procesu wchłaniania odpowiada początkowi prostoliniowego przebiegu zależności $f(t)=lnC$ i jest podstawą dzielenia wykresu zmian stężenia leku we krwi na fazę wchłaniania oraz eliminacji. Jest to jedynie umownym podziałem, gdyż w rzeczywistości procesy te zachodzą równocześnie, ale z różnymi szybkościami. 

Oprócz pokazanej powyżej sytuacji, gdy $k_a > k_e$, możliwe są jeszcze dwa przypadki [6]:
1.   **Opóźnione uwalnianie**: \
 ponownie $k_a > k_e$, ale dodatkowo występuje jeszcze $t_{lag}$, czyli czas opóźnienia wchłaniania, który występuje np. przy doustnym podaniu, kiedy to lek potrzebuje chwili, aby dostać się do docelowego miejsca,
2.   **Przedłużone uwalnianie**: \
inaczej flip-flop, $k_a < k_e$, przypadek, w którym substancja czynna jest bardzo długo uwalniana, w skrajnych przypadkach może dojść do sytuacji niepożądanej, w której dawka lecznicza nie zostanie osiągnięta.



```python
k_a = 0.1
k_e = 0.2
X_0 = 100
V_d = 50
toks = 1.5
min = 0.7

k_a1 = 1.4
k_e1 = 0.2
t_lag = 1.5

t = np.linspace(0, 40, 2000)
C = ((k_a*X_0)/((k_e - k_a)*V_d)) * (np.exp(-k_a*t) - np.exp(-k_e*t))
C1 = ((k_a1*X_0)/((k_e1 - k_a1)*V_d)) * (np.exp(-k_a1*t) - np.exp(-k_e1*t))


plt.plot(t, C, 'b', label = "przedłużone uwalnianie")
plt.plot(t + t_lag, C1, 'm', label = "opóźnione uwalnianie")
plt.axhline(toks, linestyle = '--', color = 'black')
plt.axhline(min, linestyle = '--', color = 'black')
plt.xlim(0, 40)
plt.ylim(0, 2)
plt.title("Krzywe zmian stężenia leku we krwi")
plt.xlabel("Czas [h]")
plt.xticks([5,10,15,20,25,30,35,40, t_lag], [5,10,15,20,25,30,35,40, 't_lag'])
plt.yticks([0, 1, 2, min, toks], [0, 1,  2, 'min', 'toks'])
plt.ylabel("Stężenie [mg/l]")
plt.legend(loc = "upper right")
plt.show()
```

## **Metody numeryczne**

### **Metoda Eulera**

Równania rózniczkowe można także rozwiązywać wykorzystując metody numeryczne. 
Najprostszą, a zarazem skuteczną metodą jest metoda Eulera [7]. 
Zbudujemy podstawowy program przybliżający rozwiązanie równań różniczkowych tą metodą i porównamy wyniki z wcześniej już otrzymanymi. 

Metodę podstawową Eulera używamy do równań postaci:

\begin{gather}
y' = f(y, t)
\end{gather}\
\begin{gather}
y(t_0) = y_0
\end{gather}
Korzystamy z przybliżenia:
\begin{gather}
y_{n+1} = y_n + h*f(y_n(t), t_n) 
\end{gather}
gdzie:
\begin{gather}
h =t_{n+1}-t_n
\end{gather}



```python
def Euler(f, y0, tk, h):
    t = np.arange(0, tk, h)
    y = np.zeros(len(t))
    y[0] = y0
    for n in range(0, len(t)-1):
        y[n+1] = y[n] + h * f(t[n], y[n])
    return y
```

Zobaczmy najpierw jak dokładne będzie rozwiązanie równania $\eqref{eq:rownanie1}$

```python
h = 0.1
k_e = 0.6
y0 = 50
t = np.arange(0, 10, h)
f = lambda t,y: -k_e*y
X = y0 * np.exp(-k_e*t)
plt.plot(t, Euler(f, y0, 10, h), 'b--', t, X, 'r-', markersize=1)
plt.legend(['Euler','Analitycznie'])
plt.xlim(0, 10)
plt.ylim(0, 50)
plt.title("Zmiana masy leku przy podaniu donaczyniowym, h = 0.1")
plt.xlabel("Czas [h]")
plt.ylabel("Masa leku [mg]")
plt.show()
```

Widzimy, że błąd obliczeniowy w metodzie Eulera jest tutaj znikomy, wykresy prawie idealnie na siebie nachodzą.

A co gdy zwiększymy h?

```python
h = 0.5
k_e = 0.6
y0 = 50
t = np.arange(0, 10, h)
f = lambda t,y: -k_e*y
X = y0 * np.exp(-k_e*t)
plt.plot(t, Euler(f, y0, 10, h), 'b--', t, X, 'r-', markersize=1)
plt.legend(['Euler','Analitycznie'])
plt.xlim(0, 10)
plt.ylim(0, 50)
plt.title("Zmiana masy leku przy podaniu donaczyniowym, h = 0.5")
plt.xlabel("Czas [h]")
plt.ylabel("Masa leku [mg]")
plt.show()
```

Dla $h=0.5$ rozwiązanie numeryczne nieco odbiega już od analitycznego. Zobaczyliśmy więc, że dla małych $h$ metoda Eulera jest bardzo zbliżona do metody analitycznej. Jednak jeśli $h$ jest większe to przybliżenie tą metodą jest już mniej dokładne.

Teraz rozwiązujemy równanie $\eqref{eq:rownanie5}$.

```python
h = 0.1
k_a = 1.4
k_e = 0.2
x0 = 100
t = np.arange(0,20,h)

def wzorek(t,  X):
    return k_a*x0*np.exp(-k_a*t)-k_e*X

X = ((k_a*x0)/(k_e - k_a)) * (np.exp(-k_a*t) - np.exp(-k_e*t))
plt.plot(t, Euler(wzorek, 0, 20, h), 'b--')
plt.plot( t, X, 'r-', markersize=1)
plt.legend(['Euler','Analitycznie'])
plt.xlim(0, 20)
plt.ylim(0, 100)
plt.title("Zmiana masy leku przy podaniu doustnym, h = 0.1")
plt.xlabel("Czas [h]")
plt.ylabel("Masa leku [mg]")
plt.show()
```

W tym modelu już dla $h = 0.1$ widzimy pewne niedokładności. Zobaczmy więc mniejsze $h$.

```python
h = 0.01
k_a = 1.4
k_e = 0.2
x0 = 100
t = np.arange(0,20,h)

def wzorek(t,  X):
    return k_a*x0*np.exp(-k_a*t)-k_e*X

X = ((k_a*x0)/(k_e - k_a)) * (np.exp(-k_a*t) - np.exp(-k_e*t))
plt.plot(t, Euler(wzorek, 0, 20, h), 'b--')
plt.plot( t, X, 'r-', markersize=1)
plt.legend(['Euler','Analitycznie'])
plt.xlim(0, 20)
plt.ylim(0, 100)
plt.title("Zmiana masy leku przy podaniu doustnym, h = 0.01")
plt.xlabel("Czas [h]")
plt.ylabel("Masa leku [mg]")
plt.show()
```

Widzimy, że dla $h = 0.01$ szacowanie numeryczne jest już bardzo dokładne (ciężko gołym okiem dostrzec różnicę między wykresami).
Wniosek nasuwa się taki, że metoda ta jest bardzo skuteczna przy małych wartościach $h$, a jej błąd obliczeniowy w takich przypadkach jest niewielki, więc warto ją stosować przy bardziej złożonych i skomplikowanych zagadnieniach, które ciężko rozwiązać analitycznie.

Obliczmy więc dokładnie jaki największy błąd popełniamy dla różnych wartości $h$ (od $0.01$ do $0.5$) przy równaniu $\eqref{eq:rownanie5}$. Od wartości wyliczonych analitycznie odejmujemy te wyliczone numerycznie i bierzemy największą różnicę. Następnie przedstawiamy te dane na wykresie o skali podwójnie logarytmicznej, aby znaleźć rząd zastosowanej metody Eulera, który jest równy współczynnikowi nachylenia otrzymanej prostej.


```python
k_a = 1.4
k_e = 0.2
x0 = 100

def wzorek(t,  X):
    return k_a*x0*np.exp(-k_a*t)-k_e*X

results = []
for h in np.arange(0.01, 0.5, 0.04):
  t = np.arange(0,30,h)
  X = ((k_a*x0)/(k_e - k_a)) * (np.exp(-k_a*t) - np.exp(-k_e*t))
  X1 = Euler(wzorek, 0, 30, h)
  results.append(max(abs(X1 - X)))
x = np.arange(0.01, 0.5, 0.04)
plt.loglog(x, results)
plt.grid()
plt.show()
```

Z wykresu można zauważyć, że im $h$ jest większe, tym również błąd metody jest większy, a więc tak jak wcześniej wnioskowaliśmy. Odczytać można także, że współczynnik nachylenia prostej jest równy 1, a więc rząd metody Eulera jest równy 1. 

### **Metoda Rungego-Kutty**

Inna metodą numeryczną rozwiązywania równań różniczkowych zwyczajnych jest metoda Rungego-Kutty czwartego rzędu [7]. Schemat iteracyjny ma tutaj następującą postać:

\begin{gather}
y_{n+1} = y_n+\frac{1}{6}\left(K_1+2K_2+2K_3+K_4\right)
\end{gather} gdzie:
\begin{gather}
h = t_{n+1}-t_n
\end{gather}
\begin{gather}
K_1=hf(t_n, y_n)
\end{gather}
\begin{gather}
K_2=hf(t_n+\frac{1}{2}h, y_n+\frac{1}{2}K_1)
\end{gather}
\begin{gather}
K_3=hf(t_n+\frac{1}{2}h, y_n+\frac{1}{2}K_2)
\end{gather}
\begin{gather}
K_4=hf(t_n+h, y_n+K_3)
\end{gather}


```python
def runge_kutta(f, y0, tk, h):
  t = np.arange(0, tk, h)
  y = np.zeros(len(t))
  y[0] = y0
  for n in range(0, len(t)-1):
    k1 = h * f(t[n], y[n])
    k2 = h * f(t[n]+0.5*h, y[n]+0.5*k1)
    k3 = h * f(t[n]+0.5*h, y[n]+0.5*k2)
    k4 = h * f(t[n]+h, y[n]+k3)
    y[n+1] = y[n] + 1.0/6.0 *(k1+2*k2+2*k3+k4)
  return y
```

Zastosujmy więc tę metodę do naszego modelu. Zobaczmy najpierw, co się dzieje, gdy weźmiemy $h = 0.1$ i równanie $\eqref{eq:rownanie1}$.

```python
h = 0.1
k_e = 0.6
y0 = 50
n = 0.1
t = np.arange(0, 10, n)
f = lambda t,y: -k_e*y
X = y0 * np.exp(-k_e*t)
plt.plot(t, runge_kutta(f, y0, 10, h), 'b--', t, X, 'r-', markersize=1)
plt.legend(['Runge-Kutta','Analitycznie'])
plt.xlim(0, 10)

plt.title("Zmiana masy leku przy podaniu donaczyniowym, h = 0.1")
plt.xlabel("Czas [h]")
plt.ylabel("Masa leku [mg]")
plt.show()
```

Widzimy, że metoda ta jest niemalże identyczna jak metoda analityczna. Sprawdźmy, czy także dla równania $\eqref{eq:rownanie5}$.

```python
h=0.1
k_a = 1.4
k_e = 0.2
x0 = 100
n = 0.1
t = np.arange(0,20,n)

def wzorek(t,  X):
    return k_a*x0*np.exp(-k_a*t)-k_e*X

X = ((k_a*x0)/(k_e - k_a)) * (np.exp(-k_a*t) - np.exp(-k_e*t))
plt.plot(t, runge_kutta(wzorek, 0, 20, h), 'b--')
plt.plot( t, X, 'r-', markersize=1)
plt.legend(['Runge-Kutta','Analitycznie'])
plt.xlim(0, 20)
plt.ylim(0, 100)
plt.title("Zmiana masy leku przy podaniu doustnym, h = 0.1")
plt.xlabel("Czas [h]")
plt.ylabel("Masa leku [mg]")
plt.show()
```

Jak widać, metoda Runge-Kutta jest o wiele bardziej precyzyjna niż zaprezentowana wcześniej metoda Eulera. 

Obliczmy także, jakie dokładnie błędy popełniamy dla różnych $h$ oraz wywnioskujmy stąd wcześniej już wspomniany rząd metody. 

```python
amount = []
for h in np.arange(0.01, 0.5, 0.04):
  t = np.arange(0,30,h)
  X = ((k_a*x0)/(k_e - k_a)) * (np.exp(-k_a*t) - np.exp(-k_e*t))
  X2 = runge_kutta(wzorek, 0, 30, h)
  amount.append(max(abs(X2 - X)))
x = np.arange(0.01, 0.5, 0.04)
plt.loglog(x, amount)
plt.grid()
plt.show()
```

Odczytujemy, że współczynnik nachylenia prostej jest równy 4, a więc metoda Runge-Kutta jest rzędu czwartego. Widzimy także, jak małe błędy są popełniane w tej metodzie. Jest ona niezwykle dokładna.

Ze źródła [5] zasięgneliśmy informacji odnośnie zmian stężenia APAP-u w surowicy szczura po jednorazowym podaniu leku w dawce 250mg/kg m.c. Znajdują się one w poniższej tabelce.

```python
from google.colab import files
from IPython.display import Image
uploaded = files.upload()
Image('Obraz1.png', width=800)
```

Stwórzmy wykres z tych danych.

```python
t = [0.083, 0.16, 0.25, 0.3, 0.5, 1, 1.5, 4, 12, 24]
C = [34.843, 69.573, 111.51, 114.445, 91.433, 82.197, 41.893, 20.867, 1.54, 0.14]
plt.plot(t, C, 'bo')
plt.title("Zmiana stężenia APAP-u")
plt.ylabel("Stężenie [ug/cm^3]")
plt.xlabel("Czas [h]")
plt.show()
```

Na podstawie tych danych obliczymy biologiczny okres półtrwania oraz czas, po którym lek osiągnął maksymalne stężenie. \
Aby to zrobić, musimy najpierw wyznaczyć stałą szybkości eliminacji. W tym celu czynniki przedwykładnicze związane z fazą wchłaniania i eliminacji zapisujemy jako A i B i otrzymujemy następujący wzór:

\begin{gather}
C(t) = Ae^{-k_at}-Be^{-k_et}
\end{gather}


Chcemy wyznaczyć $k_e$, więc interesują nas tylko wartości z fazy eliminacji (od $t = 0.5h$), a więc równanie ma postać:

\begin{gather}
C(t)=Be^{-k_et}
\label{eq:rownanie7} 
\tag{7}
\end{gather}

```python
from scipy.optimize import curve_fit

def func(x, B, k_e):
  return B*np.exp(-k_e * x)

popt, pcov = curve_fit(func, t[4:], C[4:])
B = popt[0]
k_e = popt[1]
t1 = np.arange(0,25)
plt.plot(t[4:], C[4:], 'bo')
plt.plot(t1, func(t1, B, k_e), 'm')
plt.title("Zmiana stężenia APAP-u")
plt.ylabel("Stężenie [ug /cm^3]")
plt.xlabel("Czas [h]")
plt.show()
print('B =',B )
print('k_e =', k_e)
```

Podstawiając wyliczone $k_e$ do wzoru na biologiczny okres półtrwania otrzymujemy:

\begin{gather}
t_{0.5} = \frac{ln2}{0.53}
\end{gather} \
\begin{gather}
t_{0.5} ≈ 1.31 \rm{h}
\end{gather}

By obliczyć czas, po którym stężenie osiągnie maksymalną wartość potrzebujemy również policzyć stałą szybkości wchłaniania. Korzystamy z metody odejmowania [6] (rzeczywiste stężenia w fazie wchłaniania odejmuje się od tych wyliczonych z równania $\eqref{eq:rownanie7}$).

\begin{gather}
C_1 - C = Ae^{-k_at}
\end{gather}

```python
C_1 = []
for i in t[0:4]:
  C_1.append(B*np.exp(-k_e * i))

różnice = []
for i in range(0,4):
  różnice.append(C_1[i] - C[i])
```

I działamy tak jak wcześniej.


```python
from scipy.optimize import curve_fit

def func2(x, A, k_a):
  return A*np.exp(-k_a * x)

popt, pcov = curve_fit(func2, t[0:4], różnice)
A = popt[0]
k_a = popt[1]
t1 = np.linspace(0,0.5,2000)
plt.plot(t[0:4], różnice, 'bo')
plt.plot(t1, func2(t1, A, k_a), 'm')
plt.title("Zmiana stężenia APAP-u")
plt.ylabel("Stężenie [ug /cm^3]")
plt.xlabel("Czas [h]")
plt.show()
print('A =',A )
print('k_a =', k_a)
```

A więc: \
\begin{gather}
t_{max} = \frac{\ln\left(\frac{14.18}{0.53}\right)}{14.18-0.53} ≈ 0.24 \rm{h}
\end{gather} \
Porównując z wartościami na wykresie, twierdzimy, że wynik jest poprawny.
